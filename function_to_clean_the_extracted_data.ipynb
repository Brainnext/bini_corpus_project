### This function is able to clean the extracted txt file but not completely

```python
import re
import os

def clean_text_from_file(file_content):
    """
    Cleans the provided text file content by removing special characters,
    formatting, and merging fragmented lines. It structures the output
    into a dictionary for easy access.
    """
    # Replace newlines with spaces to handle fragmented words across lines
    text = file_content.replace('\n', ' ')

    # Use regular expressions to find all entries
    # This pattern looks for a word, followed by a citation-like pattern
    # and then captures the description until the next entry or end of text.
    pattern = re.compile(r'(\b[A-Za-z\-]+\b)\s+\[.*?\](.*?)(?=\b[A-Za-z\-]+\b\s+\[.*?\]|$)', re.DOTALL)

    matches = pattern.finditer(text)

    cleaned_entries = {}

    for match in matches:
        word = match.group(1).strip()
        description = match.group(2).strip()

        # Merge descriptions for the same word (e.g., Udu)
        if word in cleaned_entries:
            cleaned_entries[word] += " " + description
        else:
            cleaned_entries[word] = description

    # Post-process the descriptions to remove extraneous characters and reformat
    final_output = []
    for word, desc in cleaned_entries.items():
        final_output.append(f"{word}:")

        # Heuristic splitting on ';', '.', '(', and other punctuation
        points = re.split(r';|\.\s*\(|\band\s*|\.\s*\s*|\b,', desc)

        # Special handling for "aka" varieties
        if word == 'aka':
            points_temp = []
            varieties_match = re.findall(r'aka\s+n-[\w\s\'\.]+', desc)
            if varieties_match:
                # Add varieties as individual points
                points_temp.extend(varieties_match)
                # Remove the varieties part from the original description
                desc = re.sub(r'aka\s+n-[\w\s\'\.]+', '', desc)
                # Re-split the remaining text
                points = re.split(r';|\.\s*\(|\band\s*|\.\s*\s*|\b,', desc)
                points.extend(points_temp)

        for point in points:
            point = point.strip()
            if point:
                # Remove brackets, citations, and extra spaces
                point = re.sub(r'\[.*?\]', '', point)
                point = re.sub(r'\s+', ' ', point).strip()

                # Further refine by removing leading/trailing punctuation
                point = point.strip('.,- ')

                if point:
                    final_output.append(f"  - {point}")

    return "\n".join(final_output)

def main():
    """
    Main function to ask for a file name, read its content, clean it,
    and print the cleaned output.
    """
    while True:
        file_name = input("Please enter the name of the file to clean (e.g., my_file.txt): ")

        # Check if the file exists
        if not os.path.exists(file_name):
            print(f"Error: The file '{file_name}' was not found. Please check the file name and try again.")
            continue

        try:
            with open(file_name, 'r', encoding='utf-8') as file:
                file_content = file.read()

            cleaned_output = clean_text_from_file(file_content)

            print("\n" + "="*50)
            print("Cleaned Text Output:")
            print("="*50 + "\n")
            print(cleaned_output)
            print("\n" + "="*50)
            break

        except Exception as e:
            print(f"An error occurred: {e}")
            break

if __name__ == "__main__":
    main()
```
